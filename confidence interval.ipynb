```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_excel('data-set(shoes).xlsx')
df.columns = ["January", "February", "March", "April","May","June","July","August","September","October","November","December"]
df.index=["6","6,5","7","7,5","8","8,5","9","9,5","10","10,5","11","11,5","12","13","14","15","16"]
df.head(17)
```

```python
state=pd.DataFrame()
```

```python
state["mean"]=df.mean(axis=1)
state["count"] = df.count(axis=1)
state["std"] = df.std(axis=1)
state.head()
```

```python

ci95_hi = []
ci95_lo = []

for i in state.index:
    m, c, s = state.loc[i]
    ci95_hi.append(m + 2.18*s/math.sqrt(c))
    ci95_lo.append(m - 2.18*s/math.sqrt(c))
    
state['ci95_lo'] = ci95_lo
state['ci95_hi'] = ci95_hi
state.head()
```

```python
state["last one"] = state["ci95_hi"].apply(np.ceil) 
state
```
